**1) Make a request from forecast.io for where you were born (or lived, or want to visit!).**

Tip: Once you've imported the JSON into a variable, check the timezone to make sure it's right!

Tip: How is north vs. south and east vs. west latitude/longitude represented? You sure don't use N/S and E/W!**

```python
import urllib2
response = urllib2.urlopen('http://forecast.io')
content = response.read()
url = 'https://api.forecast.io/forecast/79e09ab8f050fcb084270a66c8409501/39.9611800,-82.9987900'
response = urllib2.urlopen(url)
#get data out of response 
contents = response.read()
import json
data = json.loads(contents)
data
```

**2) What's the current wind speed?**

```python
data['currently']['windSpeed']
```

**3) The first daily forecast is the forecast for today, and the second is tomorrow, etc. For the place you decided on up above, how much of the moon will be visible tomorrow? What's will be the swing between the high and low temperatures?**

```python
data['daily']['data'][1]
```

```python
moon = data['daily']['data'][1]
moon['moonPhase']
```

**What's will be the swing between the high and low temperatures?**

```python
max = data['daily']['data'][1]
mx = max['temperatureMax']
min = data['daily']['data'][1]
mn = min['temperatureMin']
print int(mx) - int(mn)
```
