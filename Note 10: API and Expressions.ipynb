*In the terminal*

**For example:**

   *In a file with Hiliary's emails, find each time the word 'murder' is used.*

       grep murder emails.txt
    
  *To see one line an a full thought instead of a whole line of gobly gook.*

        grep -C 1 murder emails.txt
     
  *To see emails from a data or year do something like:*

        grep -C 1 2012 emails.txt 

  *If you only want emails that match the day of the week Wednesday from 2012*

        grep -E "Wednesday.*2012" emails.txt

   *And you only want emails sent in the morning:*

        grep -E "Wednesday.*2012.*AM" emails.txt
          
  *In another thought, you want to grab phone numbers mentioned in the emails.*

        grep -E "\s\d\d\d-\d\d\d\d" emails.txt 
        (for area codes)
        grep -E "\s\d\d\d-\d\d\d-\d\d\d\d" emails.txt
    
   *OR, so you don't have to type "\d" so much...:*
   
       grep -E "\d{3}-\d{3}-\d{4}" emails.txt
       
       To get just the numbers do: 
       grep -oE "\d{3}-\d{3}-\d{4}" emails.txt
       
       For number variations (if someone splits numbers with "-" or "."
       grep -oE "\d{3}[-.]\d{3}[-.]\d{4}" emails.txt

**In the toolbox:**

    "/d" - matches numbers/ matches digits 
    "/d/." - matches a number followed by a period 
    "^/d" - line start + digit (used through search/command in python)
    "^/w" - line start + 'word characters' (used through search/command in python)
    "^[A-Z]" - line start + A through Z (used through search/command in python)
    "^[A-Za-z] - line start + a-z caps/lower (used through search/command in python)
    "^\d+\.\s" - matches a line to at least one digit, + says at least more, followed by a period and some white 
    space (used in python). 
    "^\d\d?\.\s" - start, digit, period space

## Tips:

- An API ENDPOINT is basically just the URL.

- If you get an error using an API, just stick the API call URL in the browser and see what you get.
If your error says "developer inactive," probably something's wrong with your api key.

##PAGINATION:

- So APIs often don't return everything they find (If they did, an API call that has a huge result might bring down their service). Instead, they only return 10 or 20 or some number at a time. 

- So once you get the first 10 (or 20 or whatever) items, you can get the next 10 (or 20 or whatever) by telling the API to skip the first 10 (or 20 or whatever) and to give you the next 10 (or 20 or whatever).

**There are usually 2 ways you can tell an API to skip a certain number of items:**

-You specify an OFFSET, the number of items that you skip, or you specify a PAGE, the number of pages that you skip, where each page has 10 (or 20 or whatever) items.

- In the New York Times API, the Articles API makes you specify a `page` parameter and the Books API makes you specify a `offset` parameter.

## HARDCODING is when you do 
oururl = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q=Gadafi&api-key=ffaf60d7d82258e112dd4fb2b5e4e2d6:3:72421680"

**instead of something like**

ourquery = "Gadafi"
oururl = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q="+ourquery+"&api-key=ffaf60d7d82258e112dd4fb2b5e4e2d6:3:72421680"

where you have more freedom to change the value of `ourquery`. For example, you can now do something like
for ourquery in [ "Gadafi","Gaddafi","Kadafi","Qaddafi" ]:
	
    oururl = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q="+ourquery+"&api-key=ffaf60d7d82258e112dd4fb2b5e4e2d6:3:72421680"

.... So far, we've only matched exactly "11238", exactly "2012", exactly "2013", exactly "murder".
Now we want to use REGULAR EXPRESSIONS to match for looser patterns.

Let's see what lines contain "Sent:":

    - grep -E "Sent:" hillmail.txt
    -E (expression, probably) tells grep that what comes next is a regular expression.
    # We haven't used any regular expression magic yet, so this matches for exactly "Sent:".

Let's see what lines indicate that the email was sent in 2012:

    - grep -E "Sent:.*2012" hillmail.txt
     - `.` means any character
    - `*` means 0-or-more of the character that comes before 
     `.*` means 0-of-more of any character, i.e. match anything
    # `Sent:.*2012` means anything can come between `Sent:` and `2012`

Let's see what lines indicate that the email was sent on a Wednesday in 2012:

    - grep -E "Sent: Wednesday.*2012" hillmail.txt

Let's see what lines indicate that the email was sent on a Wednesday morning in 2012:

    - grep -E "Sent: Wednesday.*2012.*AM" hillmail.txt


Let's see what lines contain numbers:

     - grep -E "\d" hillmail.txt
     - `\d` means a digit

    - grep -E "\d\d" hillmail.txt
    # does NOT match lines that has a single isolated digit but matches lines with more than 2 consecutive digits.
     # It's matching exactly 2 digits, and returning the entire line, so that's why it looks like it's matching 2-or-
     more digits.

If you wanted only the parts of the lines that were matched, you can do

    - grep -oE "\d\d" hillmail.txt

Let's see what lines contain a whitespace, a digit, a whitespace:

     - grep -E "\s\d\s" hillmail.txt
    - `\s` means whitespace

Let's see what lines contain a whitespace, "murder", a whitespace:

    - grep -E "\smurder\s" hillmail.txt


Let's look for phone numbers:

Let's try the pattern 3 digits, dash, 4 digits:

    - grep -E "\d\d\d-\d\d\d\d" hillmail.txt

is the same as

    - grep -E "\d{3}-\d{4}" hillmail.txt
    - `{3}` means 3 of the character that comes before
    - So `\d{3}` means 3 digits
    -Likewise `\d{4}` means 3 digits

This actually matches stuff like "Case No. F-2015-04841", which we don't want.

Let's try the pattern whitespace, 3 digits, dash, 4 digits, whitespace:

    -grep -E "\s\d{3}-\d{4}\s" hillmail.txt
    - That got rid of lines like "Case No. F-2015-04841"
    - but misses lines like "Office: +1 202-647-9533 I Email: snipead@state.gov"

Let's try something a little broader than whitespaces:

    - grep -E "\b\d{3}-\d{4}\b" hillmail.txt
    - `\b` means word boundary, so stuff like spaces, dashes, periods, etc.

This is good! Now we also want to look for cases when the character separating the first 3 digits from the last 4 digits is a period or a space.

Let's try using . instead of - (Recall that `.` means any character)

    - grep -E "\b\d{3}.\d{4}\b" hillmail.txt
   

Let's try matching the case when there's nothing between the first 3 and the last 4 digits:

    - grep -E "\b\d{3}.?\d{4}\b" hillmail.txt
    - `?` means 0-or-1 of the character that comes before
    - So `.?` means any character or NO character

Maybe that's a bit too broad. This would actually match stuff like "123a1234" (although Hillary Clinton doesn't seem to have anything like that in her email).

Let's try narrowing it down to only a space, or a dash, or a period, or nothing between the first 3 and the last 4 digits: 

    - grep -E "\b\d{3}[ -.]?\d{4}\b" hillmail.txt
    -'[]` means any one character between `[` and `]`
    - So `[ -.]` means a space or a dash or a period
    - `.` in `[]` means an actual period, not any character


Again, we can show only the parts that match with

    -grep -oE "\b\d{3}.?\d{4}\b" hillmail.txt
    or
    - grep -o -E "\b\d{3}.?\d{4}\b" hillmail.txt
